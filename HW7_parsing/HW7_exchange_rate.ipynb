### medium: курс валют за рік лише по вівторках

```python
import datetime

import requests


# Сьогоднішня дата:

current_date = date.today()

# Знаходимо найближчий вівторок від сьогоднішньої дати:

for i in range(6):
    check_date = current_date - datetime.timedelta(i)
    if check_date.weekday() == 1:
        Tuesday = check_date
        break
print('Найближчий вівторок, що минув, був ', Tuesday)




usd_rate={}    # словник, куди будемо записувати інформацію {дата : курс}

# Запишемо курс з минулого вівторка:
params = {'date':Tuesday.strftime('%Y%m%d')}          # перетворюємо отриману дату у стрічку правильного формату
    
r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=usd&json', params).json()
r = r[0]
usd_rate[r['exchangedate']] = r['rate']
day = Tuesday

while day >= current_date - datetime.timedelta(365):    # поки дата, що перевіряється, була не більше 1 року назад
    
    day = Tuesday - datetime.timedelta(7)        #  віднімаємо 7 днів від вівторка
    params = {'date':day.strftime('%Y%m%d')}          # перетворюємо отриману дату у стрічку правильного формату
    
    r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=usd&json', params).json()
    r = r[0]
    usd_rate[r['exchangedate']] = r['rate']
    
    Tuesday = day

# Виводимо результат:
for k, v in usd_rate.items():
    print(k, v)
```
