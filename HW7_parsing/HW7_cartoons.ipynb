   
#### medium: вивести фільми та посилання на них зі всіх сторінок

```python
def list_page_cartoons(r):
    '''
        Функція для виведення назв мультиків та посилань з однієї сторінки
        
        r - request з URL потрібної сторінки
        
        Повертає:
        tuple(list_of_names, list_of_links) - кортеж з двома списками(імена, посилання)

    '''
    soup = BeautifulSoup(r.text, 'html.parser')
    cartoons = soup.find_all('a', class_='film-item')    # знаходимо <a> з класом film-item
    names = [e.get('title') for e in cartoons]            # вибираємо title зі знайдених <a>
    links = [e.get('href') for e in cartoons]             # вибираємо посилання на мультик зі знайдених <a>
    return (names, links)
```

```python
import requests
from bs4 import BeautifulSoup

link = 'https://sweet.tv/movies/multfilmi/'
r = requests.get(link, 'html')  # відкриваємо сторінку


soup = BeautifulSoup(r.text, 'html.parser')


# знаходимо загальну кількість сторінок з мультиками
# для цього шукаємо усі <li> з класом page-item, беремо назву останнього тегу

number_of_pages = int(soup.find_all('li', class_='page-item')[-1].find('a').get('title'))





# Створюємо списки для зберігання назв та лінків
names_of_cartoons = []
links_of_cartoons = []

for p in range(1,number_of_pages+1):
    
    next_page = link+'page/'+str(p)
    r = requests.get(next_page)                     # відкрваємо наступну сторінку
    names_and_links = list_page_cartoons(r)         # виклик функції, що повертає tuple(list_of_names, list_of_links)
    
    
    # Додаємо нові дані до створених списків
    names_of_cartoons.extend(names_and_links[0])
    links_of_cartoons.extend(names_and_links[1])

all_cartoons = dict(zip(names_of_cartoons, links_of_cartoons))

for k, v in all_cartoons.items():
    print(k, v)
```

### hard:  вивести фільми, посилання та акторів зі всіх сторінок

```python
# Створимо функцію-обкладинку, що розширить можлівості ф-ції  list_page_cartoons
 

def add_actors(func):
    import requests
    from bs4 import BeautifulSoup
   
    
    def wrapper(*args, **kwargs):
         
        '''
    
         повертає: 
         словник{назва (string) : посилання (string)}, 
         словник {посилання(string) : [актори (list)]}
        
        '''
            
            
        all_cartoons = func(*args, **kwargs)      # ф-ція повертає кортеж списків: ([назва],[посилання])
        
        all_cartoons = dict(zip(all_cartoons[0], all_cartoons[1]))     # створюємо словник: {назва(string): посилання(string)}
        
        cartoon_and_actors = {}       # куди будемо записувати {посилання (string): список акторів (list)}          
        
        
        for c in all_cartoons.values():     #перебираємо посилання, отримані від ф-ції list_page_cartoons
            r = requests.get(c)
            soup = BeautifulSoup(r.text, 'html.parser')
            
            tag_for_actors = soup.find_all('span', itemprop='actor')               # список тегів для кожного актора
            list_of_actors = [a.find('span').contents for a in tag_for_actors]     # список імен акторів
            
            if len(list_of_actors)==0:
                list_of_actors = ['Актори не вказані']
           
            cartoon_and_actors[c] = list_of_actors        
        
        
        # повертає: словник{назва (string) : посилання (string)}, словник {посилання(string) : [актори (list)]}
        return all_cartoons, cartoon_and_actors
    return wrapper
```

```python
import requests 
from bs4 import BeautifulSoup
@add_actors

def list_page_cartoons(r):
    '''
        Функція для виведення назв мультиків та посилань з однієї сторінки
        
        r - request з URL потрібної сторінки
        
        Повертає:
        tuple(list_of_names, list_of_links) - кортеж з двома списками(імена, посилання)

    '''
    soup = BeautifulSoup(r.text, 'html.parser')
    cartoons = soup.find_all('a', class_='film-item')     # знаходимо <a> з класом film-item
    names = [e.get('title') for e in cartoons]            # вибираємо title зі знайдених <a>
    links = [e.get('href') for e in cartoons]             # вибираємо посилання на мультик зі знайдених <a>
    return (names, links)


r = requests.get('https://sweet.tv/movies/multfilmi')

cartoons_and_links, links_and_actors = list_page_cartoons(r)

```

```python
# Використаємо створені ф-ції для кожної сторінки:

link = 'https://sweet.tv/movies/multfilmi/'
r = requests.get(link, 'html')                # відкриваємо сторінку

soup = BeautifulSoup(r.text, 'html.parser')


# знаходимо загальну кількість сторінок з мультиками
# для цього шукаємо усі <li> з класом page-item, беремо назву останнього тегу

number_of_pages = int(soup.find_all('li', class_='page-item')[-1].find('a').get('title'))

print('Всього сторінок: ', number_of_pages)


# словники, в які зберігатимемо інформацію з усіх сторінок

cartoons_and_links_all = {}
links_and_actors_all = {}

for p in range(1, number_of_pages+1):
    print('Зчитуємо дані зі сторінки №', p)
    next_page = link + 'page/' + str(p)
    r = requests.get(next_page)
    cartoons_and_links_page, links_and_actors_page = list_page_cartoons(r)
    
    # додаємо словники, отримані зі сторінки, до загальних:
    
    cartoons_and_links_all = {**cartoons_and_links_all, **cartoons_and_links_page}   
    links_and_actors_all = {**links_and_actors_all, **links_and_actors_page}
    
else:
    print('Успішно!')
```
