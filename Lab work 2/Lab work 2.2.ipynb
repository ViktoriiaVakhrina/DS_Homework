### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics # для перевірки коректності роботи написаних функцій
import math

#генеруємо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]



# Функція, що розраховує мінімальне та максимальне значення:

def min_and_max(numbers):
    '''
    Функція, що розраховує мінімальне та максимальне значення
    
    Аргументи:
    numbers - послідовність чисел (тип - list)
    
    Повертає:
        (minimum, maximum) - кортеж (перший елемент - min, другий - max). Тип даних у кортежі залежить від типу у списку numbers 
    
    '''
    minimum = numbers[0]
    maximum = numbers[0]
    for n in numbers:
        if n > maximum:
            maximum = n
        elif n < minimum:
            minimum = n
    return minimum, maximum

# Порівнюємо роботу створеної функції з вбудованими:
print('Функція для знаходження мінімального та максимального значення послідовності')
print('Вбудовані методи:', 'min:', min(list_with_random_numbers), 'max:', max(list_with_random_numbers))
print('Створений метод:', 'min:', min_and_max(list_with_random_numbers)[0], 'max:', min_and_max(list_with_random_numbers)[1])

print()



# Функція, що рохраховує середнє значення послідовності:

def mean_value(numbers):
    s = 0  # змінна для знаходження суми
    
    for n in numbers:
        s = s + n
    return s/len(numbers)

# Порівнюємо роботу створеної функції з вбудованими:
print('Функція для знаходження середнього значення послідовності')
print('Вбудований метод:', statistics.mean(list_with_random_numbers))
print('Створений метод:', mean_value (list_with_random_numbers))
    
print()



# Функція, що рохраховує медіану послідовності:

def my_median(numbers):
    numbers.sort()
    if len(numbers)%2!=0:
        return numbers[len(numbers)//2]
    else:
        return (numbers[len(numbers)//2]+numbers[(len(numbers)//2)-1])/2
    
# Порівнюємо роботу створеної функції з вбудованими:   
print('Функція для знаходження медіани послідовності')
print('Вбудований метод:', statistics.median(list_with_random_numbers))
print('Створений метод:', my_median(list_with_random_numbers)) 

print()


def add_to_dict(key_of_element, current_dict):
    '''
    Функція для додавання нового значення у словник або збільшення значення відповідного ключа на 1
    
    Аргументи:
        key_of_element - ключ, значення якого необхідно збільшити або додати нове зі значенням 1
        current_dict - словник, у який необхідно внести зміни
    '''
    if key_of_element in current_dict.keys():
        current_dict[key_of_element] += 1
    else:
        current_dict[key_of_element] = 1
        
def freq_of_numbers(numbers):
    '''
    Функція, що знаходить частоту входження кожної цифри в послідовності
    
    numbers - список чисел
    
    Повертає: 
    
    словник: ключ - цифра, значення - кількість входжень
    '''
    
    freq_of_digit = {}
    for n in list_with_random_numbers:
        if n == 100:
            add_to_dict(0, freq_of_digit)
            n=10
        else:
            add_to_dict(n%10, freq_of_digit)
            add_to_dict(n//10, freq_of_digit)
    return freq_of_digit
        
            
            
print('Частота кожної цифри в послідовності:')
print(freq_of_numbers(list_with_random_numbers))

print()


# Cереднє квадратичне відхилення 
def mean_square_dev(numbers):
    '''
    Функція, що розраховує Cереднє квадратичне відхилення 
    
    Аргументи:
    numbers - список чисел
    
    Повертає:
    значення Cередього квадратичного відхилення  (float)
    '''
    mean_of_list = mean_value(numbers)
    list_of_square_deviations = [(i-mean_of_list)**2 for i in numbers]
    # mean_square_deviation = (mean_value(list_of_square_deviations))**0.5
    mean_square_deviation = math.sqrt(mean_value(list_of_square_deviations)) # краща точність
    return mean_square_deviation
    
print('Функція для знаходження Cереднього квадратичного відхилення ')
print('Вбудований метод:', statistics.stdev(list_with_random_numbers))
print('Створений метод:', mean_square_dev(list_with_random_numbers))

```
